```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data/results/result.xlsx')
df.head()
```

```python
areas = list(df.area.unique())
areas[0] = 'ar'
dict_df = {'eligibility' : 0,
 'capability' :0,
 'available' :0,
 'protein' :0,
 'winner' :0,
 'locator' :0,
 'worlds' :0,
 'ar' :0,
 'twisted' :0,
 'lithuania' :0,
 'personnel' :0,
 'housewives' :0,
 'dialog' :0,
 'except' :0,
 'greetings ' :0}
for key in areas:
    dict_df[key] = df[df['area'].str.contains(key)]
```

```python
areas
```

```python
for key in areas:
    df = dict_df[key]
    colors = {'Кластер 0': 'red', 'Кластер 1': 'green', 'Кластер 2': 'yellow', 'Кластер 3': 'blue'}
    fig, ax = plt.subplots(figsize=(20, 20))
    for cluster_name in colors:
        color = colors[cluster_name]
        data = df[df['cluster_name'] == cluster_name]
        ax.scatter(data['x'], data['y'], color=color, linewidths=2, label=cluster_name,  edgecolors = 'black')
    for i in range(len(df)):
        ax.text(df.iloc[i,:]['x']-0.7, df.iloc[i,:]['y']+0.1, df.iloc[i,:]['keyword'][:5]+'...'+df.iloc[i,:]['keyword'][len(df.iloc[i,:]['keyword'])-5:], fontsize=15)
    ax.legend(loc='upper right', title="Ценовая категория")
    plt.title(key)
    plt.xlabel('')
    plt.ylabel('')
    plt.savefig('data/results/'+key+'.png', bbox_inches = 'tight', dpi = 100, facecolor = 'white')
```
